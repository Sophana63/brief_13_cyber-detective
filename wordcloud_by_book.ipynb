```python
import pandas as pd
import re

from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```python
def filtre(text):

    text = text.lower()
    text = text.replace('\n', ' ').replace('\r', '').replace('http', '')
    text = ' '.join(text.split())
    text = re.sub(r"[A-Za-z\.]*[0-9]+[A-Za-z%°\.]*", "", text)
    text = re.sub(r"(\s\-\s|-$)", "", text)
    text = re.sub(r"[,\!\?\%\(\)\/\"]", "", text)
    text = re.sub(r"\&\S*\s", "", text)
    text = re.sub(r"\&", "", text)
    text = re.sub(r"\+", "", text)
    text = re.sub(r"\#", "", text)
    text = re.sub(r"\$", "", text)
    text = re.sub(r"\£", "", text)
    text = re.sub(r"\%", "", text)
    text = re.sub(r"\:", "", text)
    text = re.sub(r"\@", "", text)
    text = re.sub(r"\-", "", text)
    text = re.sub(r"\;", "", text)

    return text
```

```python
df = pd.read_csv('csv/the_art_of_war.csv', delimiter='|', on_bad_lines='skip')
df.columns = ['tweet']

text = " ".join(i for i in df['tweet'].str.replace('https', ''))
clean_text = filtre(text)
clean_text
exclure_mots = ['t', 'co', 'd', 'du', 'de', 'la', 'des', 'le', 'et', 'est', 'elle', 'une', 'en', 'que', 'aux', 'qui', 'ces', 'les', 'dans', 'sur', 'l', 'un', 'pour', 'par', 'il', 'ou', 'à', 'ce', 'a', 'sont', 'cas', 'plus', 'leur', 'se', 's', 'vous', 'au', 'c', 'aussi', 'toutes', 'autre', 'comme'] + list(STOPWORDS)

# affichage avec matplotlib
wordcloud = WordCloud(collocations = False, background_color = 'white', stopwords = exclure_mots, max_words = 100).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
```
