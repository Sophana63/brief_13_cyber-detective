```python
import pyodbc
import pandas as pd
from pandasql import sqldf
```

```python
df = pd.read_csv('book.csv')

book = df[['index', 'Name', 'Price', 'Rating', 'Instock', 'Link']]
q = "SELECT * FROM book"
book = sqldf(q, globals())
#region = region[['reg_code', 'reg_nom', 'regrgp_nom']]
book
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=UTILISA-EFILC2E;'
                      'Database=books;'
                      'Trusted_Connection=yes;')

cursor = conn.cursor()
sql = "INSERT INTO book(id, title, price, rating, instock, link) VALUES(?,?,?,?,?,?)"
for i,row in book.iterrows():
    cursor.execute(sql, tuple(row))    
    conn.commit()
    
cursor.execute('SELECT * FROM book')

```
