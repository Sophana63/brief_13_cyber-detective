```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import lxml
import pandas as pd
```

```python
# création de l'une classe pouvant scraper un site web
# il faudra mettre en paramètre le lien du site à scraper
class Scraping:
    def __init__(self, link):
        self.link = link


    def myrequest (self):
        def export_table_and_print(data):
            table = pd.DataFrame(data, columns=[
                                 'Name', 'Price', 'Rating', 'Instock', 'Link'])
            table.index = table.index + 1 
        
        def get_book_attributes(book):     
            name = book.find('h3').find('a').attrs['title']
            price = book.find('p', class_='price_color').text
            price = price.replace('Â£', "")
            rating = book.find('p', class_='star-rating')
            dispo = book.find('p', class_='instock availability').text
            dispo = dispo.replace("\n\n    \n        ", "")
            dispo = dispo.replace("\n    \n", "")
            link = book.find('h3').find('a').attrs['href']
            l = rating['class']
            s = " ".join(l)
            s = s.replace("star-rating ", "")
            # remplace le texte en chiffre
            if s == "One":
                rating_number = 1
            elif s =="Two":
                rating_number = 2
            elif s =="Three":
                rating_number = 3
            elif s =="Four":
                rating_number = 4
            elif s =="Five":
                rating_number = 5
            data['Name'].append(name)
            data['Price'].append(price)
            data['Rating'].append(rating_number)
            data['Instock'].append(dispo)
            data['Link'].append(link)
            return data
    
        data = {
            'Name' : [], 'Price' : [], 'Rating' : [], 'Instock' : [], 'Link' : []
        }
        
        page = requests.get(self.link)
        if page.status_code == requests.codes.ok:
            bs = BeautifulSoup(page.text, 'lxml')
            # récupération de tous les objets
            list_all_book = bs.findAll('article', class_='product_pod')        

            for book in list_all_book:
                get_book_attributes(book)
        
        return data
```

```python
import time
from selenium.webdriver.common.by import By

# lien du site à scraper
lien = 'https://books.toscrape.com'


# ouvreture de Chrome et affiche la page demandée dans driver.get
driver = webdriver.Chrome("/usr/lib/chromium-browser/chromedriver")
driver.get("https://books.toscrape.com/")

pagination = True

# je boucle jusqu'à que le bouton Next n'existe plus
while pagination == True:
    # appelle de ma classe Scraping
    scrap = Scraping(lien)
    df = pd.DataFrame(scrap.myrequest()).reset_index()  
    print(df)
    try:  
        # récupération du nouveau lien 
        lien = driver.find_element(By.CLASS_NAME, 'next').find_element(By.CSS_SELECTOR, 'a').get_attribute("href")
        more_buttons = driver.find_element(By.CLASS_NAME, 'next').find_element(By.CSS_SELECTOR, 'a')
        driver.get(more_buttons.get_attribute('href'))        
    except:        
        pagination = False
        print("Plus de page")


```
