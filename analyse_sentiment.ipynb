```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from textblob import TextBlob
from textblob_fr import PatternTagger, PatternAnalyzer
```

```python
# lecture du CSV et transformation en dataframe
df = pd.read_csv('tweets.csv')

# je récupère toutes la colonnes des Tweets
text = df['Text']
```

```python
# fonction permettant de nettoyer les Tweets
def filtre(text):

    text = text.lower()
    text = text.replace('\n', ' ').replace('\r', '')
    text = ' '.join(text.split())
    text = re.sub(r"[A-Za-z\.]*[0-9]+[A-Za-z%°\.]*", "", text)
    text = re.sub(r"(\s\-\s|-$)", "", text)
    text = re.sub(r"[,\!\?\%\(\)\/\"]", "", text)
    text = re.sub(r"\&\S*\s", "", text)
    text = re.sub(r"\&", "", text)
    text = re.sub(r"\+", "", text)
    text = re.sub(r"\#", "", text)
    text = re.sub(r"\$", "", text)
    text = re.sub(r"\£", "", text)
    text = re.sub(r"\%", "", text)
    text = re.sub(r"\:", "", text)
    text = re.sub(r"\@", "", text)
    text = re.sub(r"\-", "", text)

    return text

# je l'applique à mes Tweets
clean_text = text.apply(filtre)
clean_text
```

```python
polarity = []
for tweet in clean_text:
  polarity.append(TextBlob(tweet).sentiment[0])

print(f"\n---------------------------------------------------------"
      "\nMaintenant que nous avons une liste avec toutes les polarités des tweets du corpus,"
      "\nnous pouvons tracer une première courbe pour voir à quoi l’évolution ressemble.")

print(f"\nOn obtient le tracé de l’évolution de la polarité des tweets.\n"
      "Comme on a 3600 tweets, cette méthode ne permet pas de voir grand chose :"
      "\n---------------------------------------------------------")
plt.plot(polarity)
```

```python
group = lambda liste, size : [liste[i:i+size] for i in range(0, len(liste), size)]

polarity_par_paquet = group(polarity,100)

liste_moyennes = []
for l in polarity_par_paquet :
  liste_moyennes.append(np.mean(l))

print(f"\n---------------------------------------------------------"
      "\nOn regroupe les tweets en paquets suivant l’ordre chronologique.\n"
      "On conservera ensuite pour chaque groupe uniquement sa moyenne.\n"
      "Je décide de les regrouper par paquet de 100 tweets."
      "\n---------------------------------------------------------")
plt.plot(liste_moyennes)
```
